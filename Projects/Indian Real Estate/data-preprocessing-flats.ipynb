```python
df = df[df['price'] != 'Price on Request']
```

```python
df.head()
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df.head()
```

```python
print(df['price_per_sqft'].head())

# Transformation function
def transform_sqft(X):
    if pd.isna(X) or X == 'nan':
        return None
    else:
      return int(str(X).replace("₹ ", "").replace("/sq.ft.", "").replace(",", ""))

df['price_per_sqft'] = df['price_per_sqft'].apply(transform_sqft)
```

```python
# Bedrooms
print(df['bedRoom'].value_counts(), end="\n##############################\n")

# Null Values
print(df['bedRoom'].isna().sum(), end="\n##############################\n")

# Fill NA values with majority
df['bedRoom'] = df['bedRoom'].fillna(3)
```

```python
# Bathrooms
print(df['bathroom'].value_counts(), end="\n##############################\n")

# Null Values
print(df['bathroom'].isna().sum(), end="\n##############################\n")

# Fill NA values with majority
df['bathroom'] = df['bathroom'].fillna(2)
```

```python
# Balconies
print(df['balcony'].value_counts(), end="\n##############################\n")

# Null Values
print(df['balcony'].isna().sum(), end="\n##############################\n")

# Fill NA values with majority
df['balcony'] = df['balcony'].fillna(3)
```

```python
def transform_text_after_whitespace(X):
    if isinstance(X, int):
        return str(X)
    else: 
        return X.split()[0]
  

df["bedRoom"] = df["bedRoom"].apply(transform_text_after_whitespace)
df["bathroom"] = df["bathroom"].apply(transform_text_after_whitespace)
df["balcony"] = df["balcony"].apply(transform_text_after_whitespace)
```

```python
# Check unique values in 'balcony' column
df['balcony'] = df['balcony'].str.strip().replace("No", 0)

# Verify the changes
print(df['balcony'].unique())
```
